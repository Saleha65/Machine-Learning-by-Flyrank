## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Question: Given a client's content inventory, which pages should a content reviewer
prioritize for refresh first?

Unit of analysis: one row = one content page (content_id), scored within its client.

Decision this improves: which pages enter this week's editorial refresh queue.

Who acts, and how: a content strategist/editor on the client's team opens a ranked list
with reason codes and works down it — updating or expanding stale content, or fixing
CTR/position issues that were flagged.

Cost of a wrong call:
- False positive (flagged but actually fine): wastes scarce editor hours on a page that
  didn't need attention.
- False negative (a real decline missed): a page that matters keeps losing visibility
  unnoticed until the next review cycle.
Because reviewer time is the scarce resource, precision at the top of the ranked list
(Precision@K) matters more than catching every declining page in the whole inventory.

Why data/ML helps: manually eyeballing tens of thousands of pages per client isn't
feasible, and no single hand rule captures the mix of staleness, visibility, position,
and CTR signals well together. The Week 1-2 notebooks already showed a learned model can
out-rank a hand-written rule at exactly this kind of "which ones first" task.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Saleha65/Machine-Learning-by-Flyrank/main/data/raw/content_refresh_anonymized.csv")

declining = df["trend_direction"].eq("down")
declining_with_demand = declining & (df["impressions_90d"] >= 100)

print(f"Total pages: {len(df):,}  across {df['client_id'].nunique()} clients")
print(f"Declining pages: {declining.sum():,} ({declining.mean()*100:.1f}%)")
print(f"Declining WITH real demand (impressions_90d >= 100): "
      f"{declining_with_demand.sum():,} ({declining_with_demand.mean()*100:.1f}%)")
print(f"Median impressions_90d across all pages: {df['impressions_90d'].median():.0f}")
```
